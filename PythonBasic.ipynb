 Q5 What is mutability in Python ?

  In Python, mutability refers to whether an object can be changed after it has been created. Objects in Python are classified as either mutable or immutable based on this characteristic.

Mutable Objects: These objects can be modified after their creation. Examples include lists, dictionaries, and sets. You can change their content, add new elements, or remove elements without creating a new object.

python
# List (Mutable)
my_list = [1, 2, 3]
my_list.append(4)  # Modifies the original list
print(my_list)  # Output: [1, 2, 3, 4]

Immutable Objects: These objects cannot be changed after their creation. Examples include strings, tuples, and integers. Any modification to these objects will result in the creation of a new object.

python
# String (Immutable)
my_string = "Hello"
new_string = my_string + " World"  # Creates a new string
print(my_string)  # Output: Hello
print(new_string)  # Output: Hello World
Understanding mutability is important for writing efficient and bug-free code, especially when dealing with data structures and functions.

Q6 Why are lists mutable, but tuples are immutable ?
# Reason Why Lists Are Mutable but Tuples Are Immutable:
Lists: In Python, lists are mutable because they are designed for situations where you need a dynamic, changeable collection. This means you can add, remove, or modify elements after the list has been created. This flexibility makes lists very powerful for various applications where data is not fixed.

Tuples: Tuples, on the other hand, are immutable. This is by design to provide a way to create fixed collections of items that cannot be changed. The immutability of tuples ensures that the data remains constant throughout the program, making them reliable for use as dictionary keys or to store data that should not change.

In essence, mutability of lists provides flexibility for dynamic data manipulation, while immutability of tuples ensures data integrity and reliability for fixed data sets.

Purpose: The == operator is used to compare the values of two objects. It checks whether the values are equal.

Purpose: The is operator is used to compare the identities of two objects. It checks whether the two references point to the same object in memory.

Usage: It is used when you need to confirm if two variables refer to the exact same object.

Example:

python
a = [1, 2, 3]
b = [1, 2, 3]

print(a is b)  # Output: False, because a and b are different objects in memory
print(a is c)  # Output: True, because c is the same object as a in memory

 Q9 What is type casting in Python ?

  Type casting, also known as type conversion, is the process of converting a value from one data type to another in Python. This can be done in two main ways: implicit casting and explicit casting.

# Implicit Casting
In implicit casting, Python automatically converts one data type to another without any explicit instruction from the programmer. This usually happens when different data types are used in an expression. For example, if you add an integer to a float, Python will automatically convert the integer to a float before performing the addition.

python
# Implicit Casting Example
x = 10   # Integer
y = 2.5  # Float
result = x + y  # Implicitly converts x to float
print(result)  # Output: 12.5
# Explicit Casting
In explicit casting, the programmer manually converts one data type to another using built-in functions. This method requires the use of specific functions like int(), float(), str(), etc. Here are some examples:

python
# Explicit Casting Examples

# Converting string to integer
num_str = "123"
num_int = int(num_str)
print(num_int)  # Output: 123

# Converting integer to float
num_float = float(num_int)
print(num_float)  # Output: 123.0

# Converting integer to string
num_str_again = str(num_int)
print(num_str_again)  # Output: "123"

   Q12 How does the elif statement work ?
    
The elif statement in Python is used to check multiple conditions in a conditional structure. It stands for "else if" and allows you to chain together several conditions, executing different blocks of code depending on which condition is met first

Here's a basic example to illustrate how the elif statement works:

python
# Example: Grading System

score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
else:
    print("Grade: F")
# How It Works:
First Condition (if):

The program checks if the score is greater than or equal to 90.

If this condition is True, it executes the block of code under this if statement and skips the rest of the elif and else statements.

Second Condition (elif):

If the first condition is False, it then checks the condition in the elif statement (whether the score is greater than or equal to 80).

If this condition is True, it executes the block of code under this elif statement and skips any remaining elif and else statements.

Subsequent Conditions (elif):

This process continues for any additional elif statements. Each condition is checked in order, and if one of the conditions is True, the corresponding block of code is executed, and the remaining conditions are skipped.

Final Condition (else):

If none of the if or elif conditions are True, the else statement block is executed. This serves as a catch-all for any cases that do not meet the previous conditions.

 Q13 What is the difference between for and while loops ?

 # for Loop
Purpose: Used to iterate over a sequence (such as a list, tuple, string, or range) for a fixed number of times or elements.

Syntax:

python
for element in sequence:
  
Use Case: When you know the number of iterations in advance, or when you're iterating over elements of a sequence.

Example:

python
# Example: Iterating over a list
fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print(fruit)
# Output: apple
#         banana
#         cherry
# while Loop
Purpose: Used to repeat a block of code as long as a specified condition is True.




while condition:
    
Use Case: When you do not know the number of iterations in advance, and you want to continue looping until a specific condition is met.

Example:

python
# Example: Looping until a condition is met
i = 1
while i <= 3:
    print(i)
    i += 1
# Output: 1
#         2
#         3
# Differences:
Iteration Control:

for loop: Iterates over each element in a sequence or runs for a specific number of times.

while loop: Repeats as long as the condition is True, which means it could potentially run indefinitely if the condition never becomes False.

Use Cases:

Use a for loop when you have a predefined sequence or a fixed number of iterations.

Use a while loop when the number of iterations depends on a condition that might change during the loop.

Initialization and Increment:

for loop: The initialization and iteration are handled automatically by the loop structure.

while loop: The initialization and increment/decrement are handled manually within the loop.
